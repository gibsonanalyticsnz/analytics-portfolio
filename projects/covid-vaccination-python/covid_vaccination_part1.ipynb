## Task 2 - 1 (Convert mixed date strings into one format: DD/MM/YYYY.)

Objective: Transform the given datetime strings into the single system format DD/MM/YYYY using both iterative and conditional control flows.
Output requirement: Print the formatted datetime strings in a list. 

```python
# # Task 2-1: Standardise dates to "DD/MM/YYYY"
# - Handle FOUR formats:
#   (A) "Month DD, YYYY"   e.g., "July 15, 2023" or "Nov 18, 2023"
#   (B) "MM-DD-YYYY"       e.g., "03-14-2024", "08-15-2023"
#   (C) "YYYY.MM.DD"       e.g., "2023.11.07"
#   (D) "DD/MM/YY"         e.g., "23/02/22"
# - Output: print list of strings in "DD/MM/YYYY"
#  if/elif branches here handle all four formats and code has try/except for long vs. short month names
# -------------------------------
from datetime import datetime  #import datetime class for parsing and formatting dates

purchase_dates = ["July 15, 2023", "2023.11.07", "03-14-2024", "23/02/22", "Nov 18, 2023", "23/04/21", "September 13, 2023", "08-15-2023"]
formatted_dates = []  # final strings in "DD/MM/YYYY"


# - Use iteration (for loop) and conditional checks (if/elif)
for s in purchase_dates:
    # (A) Month-name with comma (try full name %B, then short %b)
    if "," in s:
        try:
            dt = datetime.strptime(s, "%B %d, %Y") # %y expands to full year
        except ValueError:
            dt = datetime.strptime(s, "%b %d, %Y")
    # (B) Hyphen numeric "MM-DD-YYYY"
    elif "-" in s:
        dt = datetime.strptime(s, "%m-%d-%Y")
    # (C) Dot-separated "YYYY.MM.DD"
    elif "." in s:
        dt = datetime.strptime(s, "%Y.%m.%d")
    # (D) Slash "DD/MM/YY"
    elif "/" in s:
        dt = datetime.strptime(s, "%d/%m/%y")
    else:
        raise ValueError(f"Unrecognised date format: {s}")
    # Convert to required format
    formatted_dates.append(dt.strftime("%d/%m/%Y"))

# Print the formatted datetime strings in a list
print(formatted_dates)
```

The code for Task 2-1 standardises multiple date formats into a single system format (DD/MM/YYYY). Conditional checks using if/elif were applied to identify four input styles—month names with commas, numeric values with hyphens, year-first formats with dots, and day-first formats with slashes. Iterative control was implemented with a for loop to ensure every element in the list was processed. Conversion from text strings to date objects was carried out using datetime.strptime, followed by consistent reformatting through strftime. This approach guarantees accuracy, handles variability in input, and produces a uniform output that can be easily stored or compared in later tasks.

## Task 2 - 2 (Parse and record customer details)

Split each entry "Last,First,Rating" into last_name, first_name, and integer satisfaction, and store them in a suitable structure. A list of dictionaries was selected to preserve record order for index alignment with dates, while named fields improve readability, reduce index errors, and support future attribute extensions without restructuring.

```python
# Task 2-2(1): Extract user data and store it appropriately
user_info = ["Smith,John,5", "Johnson,Emily,3", "Brown,Michael,2", "Davis,Sarah,4", "Miller,David,5", "Wilson,Linda,5", "Moore,James,2", "Taylor,Jessica,2"]

# We'll use a list of Dictionaries.
# Each dict represents one person with named keys: last_name, first_name, satisfaction.
# To store last names, first names, and satisfaction levels in suitable data structures.

users = []
for row in user_info:
    last, first, rating = row.split(",") #Parsing - split(",") extracts last, first, and rating. Rating is converted into int
    users.append({
        "last_name": last.strip(),
        "first_name": first.strip(),
        "satisfaction": int(rating.strip())
    })

# Print to verify the stored data
print(users)

 
```


In Task 2-2, customer information in the form "Last,First,Rating" was parsed into distinct attributes. The values were split into last name, first name, and satisfaction score, with satisfaction converted into an integer for later analysis. A list of dictionaries was selected as the storage structure because it offers named fields, which make the code more readable and reduce indexing errors. Using a list preserves the original order, allowing alignment with purchase dates in Task 2-1, while the dictionary format allows straightforward extension with new fields if required. This combination ensures flexibility, clarity, and efficiency for both processing and reporting.

## Task 2 - 3 (combine into a single nested structure)

Combine the outputs of Task 2-1 (the list of formatted dates) and Task 2-2 (the list of user dictionaries) into one nested structure so each record contains a customer’s name, satisfaction, and first_use_date.

```python
# Task 2-3: Combine outputs from Task 2-1 (formatted_dates) and Task 2-2 (users)
# Objective: Build a single nested structure where each record contains
# last_name, first_name, satisfaction, and first_use_date.
# If the list are of different length, this code assumes equal lengths of users and dates

# Step 1: Create a new combined list
combined_records = []

# Step 2: Use a for loop with index (lab style) to merge user info with corresponding date
for i in range(len(users)):  # loop merges records from both lists into one dictionary per person.
    record = {
        "last_name": users[i]["last_name"],
        "first_name": users[i]["first_name"],
        "satisfaction": users[i]["satisfaction"],
        "first_use_date": formatted_dates[i]
    }
    combined_records.append(record)

# Step 3: Print the final nested structure
for rec in combined_records:
    print(rec)
```

Task 2-3 integrated the outputs from the previous tasks into a single nested structure. An index-based loop aligned each customer record with the corresponding date by position, creating a unified dictionary containing last name, first name, satisfaction score, and first-use date. The final list of dictionaries stores all relevant details for each customer together, removing the need to cross-reference separate lists. This structure makes it simple to identify individuals, track when they first used the service, and evaluate their satisfaction level. It also supports filtering (e.g., by satisfaction), sorting (e.g., by date), and exporting for further analysis. The chosen design therefore improves both usability and analytical value of the dataset.
